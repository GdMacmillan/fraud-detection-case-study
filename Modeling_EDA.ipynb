```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df = pd.read_json('data/data.json')

df['fraud'] = 0
def is_fraud(x):
    if x == 'fraudster' or x == 'fraudster_event':
        return 1
    else:
        return 0

df['fraud'] = df['acct_type'].apply(is_fraud, 1)

def is_listed(x):
    if x == 'y':
        return 1
    else:
        return 0

df['listed'] = df['listed'].apply(is_listed, 1)

df = df[df['sale_duration'] >= 0] # drop outlier case where sale duration is negative

def is_nan(x):
    if np.isnan(x):
        return 1
    else:
        return 0

df['has_header_nan'] = df['has_header'].apply(is_nan, 1)
df['has_header'].fillna(0, inplace=True)

# df has heavily imbalanced classes with approx. 10% being fraud

# all fraud cases
df_fraud = df[df.fraud == 1]
# random sample not fraud
df_not_fraud = df[df.fraud == 0].sample(df_fraud.shape[0])
# create new balanced class dataframe
df = pd.concat([df_fraud, df_not_fraud], ignore_index=True)
df.head()
```

```python
from statsmodels.discrete.discrete_model import Logit
from statsmodels.tools import add_constant

train_cols = ['channels', 'body_length', 'gts', 'user_age', 'sale_duration', 'name_length']
X = df[train_cols].values
X_const = add_constant(X, prepend=True)
y = df['fraud'].values
logit = Logit(y, X_const)
logit_model = logit.fit()
```

```python
print logit_model.summary()
```

```python
import numpy as np
from sklearn.cross_validation import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score

kfold = KFold(len(y))

accuracies = []
precisions = []
recalls = []

for train_index, test_index in kfold:
    model = LogisticRegression()
    model.fit(X[train_index], y[train_index])
    y_predict = model.predict(X[test_index])
    y_true = y[test_index]
    accuracies.append(accuracy_score(y_true, y_predict))
    precisions.append(precision_score(y_true, y_predict))
    recalls.append(recall_score(y_true, y_predict))

print "accuracy:", np.average(accuracies)
print "precision:", np.average(precisions)
print "recall:", np.average(recalls)
```

```python
from sklearn.metrics import roc_curve
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
model = LogisticRegression()
model.fit(X_train, y_train)
probabilities = model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, probabilities)
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate (1 - Specificity)")
plt.ylabel("True Positive Rate (Sensitivity, Recall)")
plt.title("ROC plot of fraud data")
plt.show()
```

Now I figure it may be important to include a feature that would be the timedelta between when the user was created and when the event was created. If a fraudster is creating a fraudulent event with the intention of scamming people into purchasing tickets to a fake event, then maybe there is some correlation between the probability of the event being fraud and the amount of time between creating their account and when the event was created.  

```python
import datetime

def days_between_timestamp(x):
    return abs(datetime.datetime.fromtimestamp(x[0]) - (datetime.datetime.fromtimestamp(x[1]))).days

df['days_user_event'] = df[['event_created', 'user_created']].apply(days_between_timestamp, 1)
df.boxplot(column='days_user_event', by='fraud')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error

X = df[train_cols].values
y = df['fraud'].values
X_train, X_test, y_train, y_test = train_test_split(X, y)

rf = RandomForestClassifier(n_estimators=100, oob_score=True)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print 'RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred))
print 'OOB_error: ', rf.oob_score_
indx = np.arange(len(train_cols))
srt = np.argsort(rf.feature_importances_)[::-1]
feat_names = [str(s) for s in np.array(train_cols)[srt]]
plt.bar(indx, rf.feature_importances_[srt])
plt.xticks(indx, feat_names, rotation=45)
plt.ylabel("Normalized Importances")
plt.title("Feature Importances for Rand. For.")
plt.tight_layout()
plt.show()
```
